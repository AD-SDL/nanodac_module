# nanodac Interface Tests

Direct tests of the `Nanodac` driver in `src/nanodac_interface.py` (pure Modbus/TCP, no MADSci).
Requires a live nanodac reachable on the network (default here: 9-BM, unit id 1).

```python
from nanodac_interface import Nanodac
```

```python
nanodac = Nanodac("192.168.1.100")  # replace with your controller's IP (Modbus/TCP, unit id 1)
nanodac.get_status()
```

```python
nanodac.get_temperature()  # Loop.1 process value (measured temperature)
```

```python
nanodac.get_target_temperature()  # target setpoint (None if never written)
```

```python
nanodac.get_working_setpoint()  # working setpoint (ramps toward the target)
```

```python
nanodac.get_output()  # active control output (%)
```

```python
nanodac.get_mode()  # "Auto" or "Manual"
```

```python
# Every getter accepts an explicit loop (1 or 2); default is loop 1.
nanodac.get_status(loop=1)
```
