## Read Actions

```python
# get_temperature -> {'temperature': <PV>}
request = ActionRequest(action_name="get_temperature", args={})
response = client.send_action(action_request=request)
```

```python
client.get_action_result(response.action_id)
```

```python
# Actions accept an optional loop arg (defaults to the node's configured loop).
request = ActionRequest(action_name="get_temperature", args={"loop": 1})
response = client.send_action(action_request=request)
```

```python
client.get_action_result(response.action_id)
```

```python
# get_setpoint -> {'setpoint': <target SP or None>}
request = ActionRequest(action_name="get_setpoint", args={})
response = client.send_action(action_request=request)
```

```python
client.get_action_result(response.action_id)
```

```python
# get_output -> {'output': <control output %>}
request = ActionRequest(action_name="get_output", args={})
response = client.send_action(action_request=request)
```

```python
client.get_action_result(response.action_id)
```
